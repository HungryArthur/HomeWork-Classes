```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return -0.5 * x + 1.2



x0 = 0.5 # Начальное приближение
n_iter = 10 # Количество итераций

# Списки для хранения точек итерационного процесса
x_vals = [x0]
y_vals = [f(x0)]

# Итерационный процесс: x_{n+1} = f(x_n)
x = x0
for i in range(n_iter):
    x_new = f(x)
    x_vals.append(x_new)
    y_vals.append(f(x_new))
    x = x_new

# Построение графиков
plt.figure(figsize=(10, 8))

# Линия y = x (биссектриса)
x_line = np.linspace(0, 2, 100)
plt.plot(x_line, x_line, 'k--', label='y = x', alpha=0.7)

# График функции f(x)
x_f = np.linspace(0, 2, 100)
plt.plot(x_f, f(x_f), 'b-', label='f(x) = -0.5x + 1.2', linewidth=2)

# Итерационный процесс (лестница)
for i in range(n_iter):
    # Вертикальная линия от (x_i, x_i) до (x_i, f(x_i))
    plt.plot([x_vals[i], x_vals[i]], [x_vals[i], y_vals[i]], 'r-', linewidth=1.5, alpha=0.7)
    # Горизонтальная линия от (x_i, f(x_i)) до (f(x_i), f(x_i))
    plt.plot([x_vals[i], y_vals[i]], [y_vals[i], y_vals[i]], 'g-', linewidth=1.5, alpha=0.7)

# Точки итераций
plt.plot(x_vals, y_vals, 'ro', markersize=8, label='Итерационные точки')

# Отметим неподвижную точку
x_fixed = 0.8
plt.plot(x_fixed, x_fixed, 'k*', markersize=15, label='Неподвижная точка')

# Настройки графика
plt.xlim(0, 1.5)
plt.ylim(0, 1.5)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.title('Сходящийся итерационный процесс при -1 < f\'(x) < 0', fontsize=16)
plt.legend()
plt.grid(True, alpha=0.3)
plt.gca().set_aspect('equal', adjustable='box')

# Вывод значений итераций
print("Итерационный процесс:")
for i, x_val in enumerate(x_vals):
    print(f"x{i} = {x_val:.6f}")

plt.show()
```
