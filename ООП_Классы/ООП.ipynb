#### Задание E

<img src="img/1_E.png" width="600">


```python
from random import choice

class Parrot:
	def __init__(self, _s, _num = 1):
		self.s = _s
		self.num = _num
		self.mozg = [_s]

	def say(self, _num = 1):
		self.num = _num
		self.s = choice(self.mozg)
		print(self.s * self.num)
	
	def learn(self, _s):
		self.s = _s
		self.mozg.append(self.s)

p = Parrot("Привет")
p.say()
p.learn("Гав! ")
p.say()
p.say(3)
```

```python
class LampRow:
	def __init__(self, _count = "8"):
		self.count = int(_count)
		self._state = "0" * int(_count)
	
	def get_state(self):
		return self._state
	
	def set_state(self, _state):
		if len(_state) == self.count:
			self._state = _state
		else:
			print(f"ошибка")
			self._state = "0" * self.count
	
	state = property(get_state, set_state)
	
	def show(self):
		result = ""
		for i in self._state:
			if i == "1":
				result += "*"
			elif i == "2":
				result += "o"
			else:
				result += "-"
		print(result)

# теперь можно писать как и число так и строку
lamps = LampRow(6)
lamps = LampRow("6")
lamps.show()              # ------
lamps.state = "102102"
print(lamps.state)        # 102102
lamps.show()              # *-o*-o
lamps.state = "10201010"  # ошибка
print(lamps.state)        # 000000
lamps.show()              # ------
```
